## 3.2 Problem - 2 - Some Advance Data Exploration Task:

### Task - 1 - Setup Task - Preparing the South-Asia Dataset:

```python
south_asian_countries = ["Afghanistan", "Bangladesh", "Bhutan", "India","Maldives", "Nepal", "Pakistan", "Sri Lanka"]
matching_countries = df[df['Country name'].isin(south_asian_countries)]
matching_countries.to_csv('south_asian_filtered.csv', index=False)

```

## Composite score ranking

```python
sa = pd.read_csv('south_asian_filtered.csv')
def composite_score(row):
        return  (0.40 * row['Log GDP per capita'] + 0.30 * row['Social support'] + 0.30 * row['Healthy life expectancy'])
    
sa['Composite score'] =sa.apply(composite_score, axis=1)
sa
```

```python
descend_asian = sa.sort_values(by='Composite score',ascending=False)
descend_asian
```

```python
top_five_asian_countries=descend_asian.head(5)
plt.figure(figsize=(5,3))
plt.barh(top_five_asian_countries['Country name'], top_five_asian_countries['Composite score'], color='skyblue', edgecolor='black')
plt.xlabel('Composite Score', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.title('Top 5 Countries by Composite Score', fontsize=14)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 4))
x = range(len(sa))
plt.bar(x, sa['score'], width=0.4, label='Original Score', align='center', color='orange')
plt.bar(x, sa['Composite score'], width=0.4, label='Composite score', align='edge', color='purple')
plt.xlabel('Countries (Ordered by Composite Score)')
plt.ylabel('Scores')
plt.title('Comparison of Original Score vs Composite Score')
plt.xticks(x, sa['Country name'])
plt.legend()
plt.tight_layout()
plt.show()
```

```python
gdp_q1 = sa['Log GDP per capita'].quantile(0.25)
gdp_q3 = sa['Log GDP per capita'].quantile(0.75)
gdp_iqr = gdp_q3 - gdp_q1


score_q1 = sa['score'].quantile(0.25)
score_q3 = sa['score'].quantile(0.75)
score_iqr = score_q3 - score_q1


gdp_lower = gdp_q1 - 1.5 * gdp_iqr
gdp_upper = gdp_q3 + 1.5 * gdp_iqr
score_lower = score_q1 - 1.5 * score_iqr
score_upper = score_q3 + 1.5 * score_iqr


sa['Outlier'] = ((sa['Log GDP per capita'] < gdp_lower) | (sa['Log GDP per capita'] > gdp_upper) |
                        (sa['score'] < score_lower) | (sa['score'] > score_upper))

plt.figure(figsize=(10, 6))
plt.scatter(sa['Log GDP per capita'], sa['score'], label='Normal Countries', color='skyblue')


outliers = sa[sa['Outlier']]
plt.scatter(outliers['Log GDP per capita'], outliers['score'], label='Outliers', color='red')

# Add labels and title
plt.xlabel('GDP per Capita')
plt.ylabel('Score')
plt.title('Scatter Plot of GDP per Capita vs. Score (Highlighting Outliers)')
plt.legend()
plt.show()
```

```python
south_asian_freedom_corr = sa['Freedom to make life choices'].corr(sa['score'], method='pearson')
south_asian_generosity_corr = sa['Generosity'].corr(sa['score'], method='pearson')
print(f"Correlation between Freedom to Make Life Choices and Score: {south_asian_freedom_corr:.2f}")
print(f"Correlation between Generosity and Score: {south_asian_generosity_corr:.2f}")
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='Freedom to make life choices', y='score', data=sa, 
            scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
plt.title('Freedom to make life choices vs. Score')
plt.xlabel('Freedom to Make Life Choices')
plt.ylabel('Score')
plt.show()

# Plot Generosity vs. Score
plt.figure(figsize=(10, 6))
sns.regplot(x='Generosity', y='score', data=sa, 
            scatter_kws={'color': 'green'}, line_kws={'color': 'orange'})
plt.title('Generosity vs. Score')
plt.xlabel('Generosity')
plt.ylabel('Score')
plt.show()
```

```python
sa['GDP-Score Gap'] = sa['Log GDP per capita'] - sa['score']
sa
```

```python
ascend_GDP_GAP = sa.sort_values(by='GDP-Score Gap',ascending=True)
descend_GDP_GAP = sa.sort_values(by='GDP-Score Gap', ascending=False)
positive_gaps = descend_GDP_GAP.head(3)
positive_gaps
```

```python
negative_gaps = ascend_GDP_GAP.head(3)
negative_gaps
```

```python
top_gaps = pd.concat([positive_gaps, negative_gaps])
plt.figure(figsize=(10, 6))
plt.bar(top_gaps['Country name'], top_gaps['GDP-Score Gap'], color=['green' if x > 0 else 'red' for x in top_gaps['GDP-Score Gap']])
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  
plt.title('Top 3 Largest Positive and Negative GDP-Score Gaps')
plt.xlabel('Country')
plt.ylabel('GDP-Score Gap')
plt.xticks()
plt.show()
```

```python
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers


south_asia_score_outliers = detect_outliers(sa, 'score')
south_asia_gdp_outliers = detect_outliers(sa, 'Log GDP per capita')
middle_east_score_outliers = detect_outliers(me, 'score')
middle_east_gdp_outliers = detect_outliers(me, 'Log GDP per capita')
south_asia_outliers = pd.concat([south_asia_score_outliers[['Country name', 'score', 'Log GDP per capita']],
                                 south_asia_gdp_outliers[['Country name', 'score', 'Log GDP per capita']]])

middle_east_outliers = pd.concat([middle_east_score_outliers[['Country name', 'score', 'Log GDP per capita']],
                                  middle_east_gdp_outliers[['Country name', 'score', 'Log GDP per capita']]])
print("South Asia Outliers:\n", south_asia_outliers)
print("Middle East Outliers:\n", middle_east_outliers)
```
